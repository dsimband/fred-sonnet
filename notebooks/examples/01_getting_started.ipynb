# Getting Started with FRED API

This notebook demonstrates basic usage of the FRED API client. We'll cover:
- Setting up the client
- Fetching basic economic data
- Converting responses to pandas DataFrames

```python
import sys
sys.path.append('../..')

from src.api.fred_client import FREDClient
from src.models.economic_data import Series
from src.utils.helpers import series_to_dataframe

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Initialize the FRED Client

First, we'll create an instance of our FRED client. Make sure you have your API key in the `.env` file.

```python
client = FREDClient()
```

## Fetch GDP Data

Let's start by fetching GDP data as an example.

```python
# Fetch GDP data
gdp_data = client.get_series('GDP')
gdp_series = Series.from_api_response(gdp_data)
gdp_df = series_to_dataframe(gdp_series)

# Display the first few rows
gdp_df.head()
```

## Basic Visualization

Let's create a simple plot of the GDP data.

```python
plt.figure(figsize=(12, 6))
plt.plot(gdp_df.index, gdp_df['value'])
plt.title('US GDP Over Time')
plt.xlabel('Date')
plt.ylabel('Billions of Dollars')
plt.grid(True)
plt.show()
```
